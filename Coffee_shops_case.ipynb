```python
competition_shops=['Bakery','Café','Coffee Shop','Cafe, Coffee, and Tea House','Tea Room','Snack Place','Donut Shop','Sandwich Spot']
```

```python
competition_shop_count = hyd_venues.groupby("Area")["Venue Category"].apply(lambda x: ((x=='Bakery') | (x=='Café') | (x=='Coffee Shop') | (x=='Cafe, Coffee, and Tea House') | (x=='Tea Room') | (x=='Snack Place') | (x=='Donut Shop') | (x=='Sandwich Spot')).sum()).reset_index(name="Competition_shops_count")
```

```python
print(competition_shop_count)
```

```python
competition_shop_count_sorted = competition_shop_count.sort_values(by="Competition_shops_count")
competition_shop_count_sorted
```
